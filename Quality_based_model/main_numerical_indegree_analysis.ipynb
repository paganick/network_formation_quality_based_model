# Numerical Indegree Analysis

In what follows, we run 1000 simulations with 1000 nodes. We then study the in-degree distribution in terms of:
1) probability density function;
2) complementary cumulative distribution function;
3) probability density function as a function of the quality ranking.

We also compare three different scenarios for the meeting distribution: 1) uniform random, 2) pure Preferential Attachment based, 3) Mixed.

The results are reported in Supplementary Fig. 1, left column. 

### Extreme PA

```python
np.random.seed(0)
plt.rcParams["figure.figsize"]=[10,7]
tt=TicToc()
tt.tic()
N = [1000]
nsim = 10000
save_results = True
model_verbose = False
model_folder = 'numerical_indegree_analysis_extremePA_1000/'
model_parameters = Model_parameters(N, nsim, save_results, model_folder, model_verbose)

T = 10
pPA = -1.0
speed_up = True
simulation_verbose = False
simulation_parameters = Simulation_parameters(T, pPA, speed_up, simulation_verbose)

model = Model(model_parameters, simulation_parameters)
model.run()
model.numerical_indegree_analysis()
tt.toc()
```
